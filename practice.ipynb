```python
def to_jaden_case(string):
    return " ".join([c.capitalize() for c in string.split()])

print(to_jaden_case("hello world time to capitalize"))
```

```python
def high_and_low(numbers):   
    list1 = [int(i) for i in numbers.split()]
    return str(max(list1)) + " " + str(min(list1))

print(high_and_low("8 3 -5 42 -1 0 0 -9 4 7 4 -4"))
print(high_and_low("1 2 3"))
```
