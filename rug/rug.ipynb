```python
from PIL import Image, ImageDraw

# Set the width and height of the image
width = 5000
height = 5000

# Create a new image with the specified dimensions and a white background
image = Image.new('RGB', (width, height), 'white')

# Create a drawing object to draw on the image
draw = ImageDraw.Draw(image)

# Set the color for the pixels to be set
red = (255, 0, 0)

# Iterate through the grid of pixels
for i in range(width):
    for j in range(height):
        x, y = i, j
        while x > 0 and y > 0:
            if x % 3 == 1 and y % 3 == 1:
                break
            x = int(x/3)
            y = int(y/3)
        else:
            draw.point((i, j), fill=red)

# # Save the image to a file
# image.save('grid_of_red_pixels.png')

# Display the image
image.show()
```

```python
import numpy as np
from PIL import Image

def apply_rule(x, y):
    while x > 0 and y > 0:
        if x % 3 == 1 and y % 3 == 1:
            return False
        x, y = x // 3, y // 3
    return True

# Set the width and height of the image
width = 5000
height = 5000

# Create a NumPy array to represent the image grid
image_grid = np.zeros((height, width), dtype=bool)

# Iterate through the grid of pixels
for i in range(width):
    for j in range(height):
        image_grid[j, i] = apply_rule(i, j)

# Create a new image with the specified dimensions and a white background
image = Image.new('1', (width, height), 'white')

# Copy the image grid data to the image
image.putdata(image_grid.flatten())

# Display the image
image.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sierpinski_carpet(depth):

    if depth == 0:
        return np.ones((1, 1))

    carpet = sierpinski_carpet(depth - 1)
    center = np.zeros_like(carpet)
    
    squares = np.block([[carpet, carpet, carpet],
                        [carpet, center, carpet],
                        [carpet, carpet, carpet]])

    return squares


fig, ax = plt.subplots(figsize=(10, 10))

ax.set_aspect('equal')
ax.set_axis_off()

carpet = sierpinski_carpet(depth=8)

plt.imshow(carpet, cmap='binary')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def fill_square(x1, y1, size, img):
    for x in range(x1, x1+size):
        for y in range(y1, y1+size):
            img[x, y] = 0
    return img


levels = 8
size = 3**levels
# size = 5000

img = np.ones((size, size), dtype=np.uint8)

for level in range(1, levels+1):
    square_size = int(size/(3**level))
    for x in range(0, 3**level, 3):
        x = int((x+1)*square_size)
        for y in range(0, 3**level, 3):
            y = int((y+1)*square_size)
            img = fill_square(x, y, square_size, img)


plt.axis('off')
plt.imshow(img, cmap='binary')
# plt.imsave('sierpinski.png', img, cmap='binary')
plt.show()
```
