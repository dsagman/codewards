```python
import math 
# import tqdm to print results
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python


def count_ways_to_sum(n):
    global triangle_numbers
    ways = set()
    max_triangle_number = math.ceil(math.sqrt(8*n + 1) - 1) // 2
    # Check for sums of three triangular numbers
    for i in range(max_triangle_number + 1):
        for j in range(i, max_triangle_number + 1):
            for k in range(j, max_triangle_number + 1):
                if triangle_numbers[i] + triangle_numbers[j] + triangle_numbers[k] == n:
                    ways.add((triangle_numbers[i], triangle_numbers[j], triangle_numbers[k]))
                    break
    return len(ways)


limit = 10000
triangle_numbers = [n * (n + 1) // 2 for n in range(limit)]
results = []

for n in tqdm(range(1, limit + 1)):
    count = count_ways_to_sum(n)
    results.append(count)

print(results)

```

```python
# plot the results
plt.plot(range(1, limit + 1), results, 'o', markersize=2)
plt.xlabel('n')
plt.ylabel('Number of ways to write n as a sum of three triangular numbers')
plt.show()
```
